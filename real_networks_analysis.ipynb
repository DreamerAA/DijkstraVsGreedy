```python
import pandas as pd
import networkx as nx
import xarray as xr
import numpy as np
```

```python
file_name = "./ego_facebook_result_6.nc"
df = xr.open_dataset(file_name)
```

```python
df.ln10_z.plot()
```

```python
coord_names = ["ln10_u", "ln2_p"]
mdf = xr.Dataset({
    "ln10_z": (coord_names, df.ln10_z.data),
},
    coords={
    "ln10_u": df.ln10_u,
    "ln2_p": df.ln2_p,
})
```

```python
mdf.ln10_z.plot()
```
